```python
def fun5(a):
    if (str(a['LOCAL_PATIENT_IDENTIFIER']).startswith('L_') ):
        return 'l'
    if (str(a['LOCAL_PATIENT_IDENTIFIER']).startswith('R_') ):
        return 'r'
def fun6(b):
    if (str(b['MDT DISCUSSION INDICATOR']).startswith('N_') ):
        return 'n'
    if (str(b['MDT DISCUSSION INDICATOR']).startswith('Y_') ):
        return 'y'
def fun7(c):
    if (str(c['RECURRENCE INDICATOR']).startswith('N_No') ):
        return 'n_n'
    if (str(c['RECURRENCE INDICATOR']).startswith('Y_Yes') ):
        return 'y_y'
def fun8(d):
    if (str(d['CANCER CARE PLAN INTENT']).startswith('P_P') ):
        return 'pal'
    if (str(d['CANCER CARE PLAN INTENT']).startswith('N_N') ):
        return 'no'
    if (str(d['CANCER CARE PLAN INTENT']).startswith('C_C') ):
        return 'cur'
    if (str(d['CANCER CARE PLAN INTENT']).startswith('S_S') ):
        return 'suptve'
def fun9(e):
    if (str(e['PLANNED CANCER TREATMENT TYPE']).startswith('3_C') ):
        return 'curatve'
    if (str(e['PLANNED CANCER TREATMENT TYPE']).startswith('1_S') ):
        return 'surg'
    if (str(e['PLANNED CANCER TREATMENT TYPE']).startswith('2_T') ):
        return 'tel'
    if (str(e['PLANNED CANCER TREATMENT TYPE']).startswith('4_H') ):
        return 'harm'
    if (str(e['PLANNED CANCER TREATMENT TYPE']).startswith('5_S') ):
        return 'sup'
    if (str(e['PLANNED CANCER TREATMENT TYPE']).startswith('6_B') ):
        return 'brach'
    if (str(e['PLANNED CANCER TREATMENT TYPE']).startswith('7_B') ):
        return 'bio'
def fun10(f):
    if (str(f['PERFORMANCE STATUS (ADULT)']).startswith('1_A') ):
        return 'a'
    if (str(f['PERFORMANCE STATUS (ADULT)']).startswith('2_R') ):
        return 'res'
    if (str(f['PERFORMANCE STATUS (ADULT)']).startswith('3_A') ):
        return 'able'
    if (str(f['PERFORMANCE STATUS (ADULT)']).startswith('4_C') ):
        return 'cap'
    if (str(f['PERFORMANCE STATUS (ADULT)']).startswith('5_C') ):
        return 'complete'
def fun11(g):
    if (str(g['NO CANCER TREATMENT REASON']).startswith('01_P') ):
        return 'unfit1'
    if (str(g['NO CANCER TREATMENT REASON']).startswith('02_U') ):
        return 'unfit2'
    if (str(g['NO CANCER TREATMENT REASON']).startswith('03_U') ):
        return 'unfit3'
    if (str(g['NO CANCER TREATMENT REASON']).startswith('04_U') ):
        return 'unfit4'
    if (str(g['NO CANCER TREATMENT REASON']).startswith('05_U') ):
        return 'unfit5'
    if (str(g['NO CANCER TREATMENT REASON']).startswith('06_D') ):
        return 'unfit6'
    if (str(g['NO CANCER TREATMENT REASON']).startswith('07_N') ):
        return 'unfit7'
    if (str(g['NO CANCER TREATMENT REASON']).startswith('08_O') ):
        return 'unfit8'
    if (str(g['NO CANCER TREATMENT REASON']).startswith('01_P') ):
        return 'unift9'
```
