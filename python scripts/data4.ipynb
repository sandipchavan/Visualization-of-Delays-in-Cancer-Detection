```python
import pandas as pd
from pandas import DataFrame as df
import random
import datetime
```

```python
df1=pd.read_excel('cancer_data.xlsx',sheet_name="DEMOGRAPHICS")
```

```python
df2=pd.read_excel('cancer_data.xlsx',sheet_name="REFERRAL")
df2.head()
```

```python
lst=["14_Suspected sarcomas",
"03_Suspected lung cancer","04_Suspected head and neck cancers",'15_Other suspected cancer']*900
sam=random.sample(lst,900)
lst=["14_Suspected sarcomas","01_Suspected breast cancer",
"03_Suspected lung cancer",'15_Other suspected cancer']*900
lst_sm=random.sample(lst,900)
```

```python
def fun1(x):
    if str(x["GENDER"]).startswith("F"):
        return lst_sm[random.randint(0,800)]
    else:
        return sam[random.randint(0,800)]
```

```python
temp=pd.concat([df1.GENDER,df2],axis=1)
temp.head()
```

```python
df2["URGENT CANCER REFERRAL TYPE"]=temp.apply(fun1,axis=1)
```

```python
df2.head()
```

```python
lst1=["1_suspected cancer",
"3_no new cancer diagnosis identified by the Trust",
'5_diagnosis of new cancer confirmed - treatment not yet planned',
'6_diagnosis of new cancer confirmed - treatment planned',
'7_diagnosis of new cancer confirmed - no treatment planned'
,'1_suspected cancer'
'1_suspected cancer'
]*900
lst_sample=random.sample(lst1,900)
lst_sample[0]
```

```python
df2["CANCER STATUS"]=df2.apply(lambda x:lst_sample[random.randint(1,100)],axis=1)
```

```python
df3=pd.read_excel("cancer_data.xlsx",sheet_name="IMAGING")
```

```python
writer = pd.ExcelWriter('cancer_data2.xlsx', engine='xlsxwriter')
df1.to_excel(writer,sheet_name="DEMOGRAPHICS",index=None)
df2.to_excel(writer,sheet_name='REFERRAL',index=None)
df3.to_excel(writer,sheet_name='IMAGING',index=None)
writer.save()
```
