```python
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df3=pd.read_excel('cancer_data6 - Copy.xlsx',sheet_name='IMAGING')
```

```python
string_lst=['ANATOMICAL EXAMINATION SITE','INVASIVE LESION SIZE (RADIOLOGICAL DETERMINATION)','CANCER IMAGING MODALITY']
```

```python
with sns.plotting_context(font_scale=2,rc={'xtick.major.pad': 5.0,'xtick.major.width':0.7,'xtick.minor.width': 0.4,'figure.figsize': [5.0, 5.0],'legend.fontsize': 20,'xtick.labelsize': 10.0}):
    for i in range(0,len(string_lst)+1):
        s1=sns.pairplot(df3,size=4,kind="scatter",hue=string_lst[i],palette=sns.hls_palette(i+2, l=.4))
        s1.savefig("img"+str(i)+".png")
```
