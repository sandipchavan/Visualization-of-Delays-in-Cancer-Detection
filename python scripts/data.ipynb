```python
def fun2(x):
    if (str(x['LOCAL_PATIENT_IDENTIFIER']).startswith('OD_') )  and ( str(x['statename']).startswith("KA") or str(x["statename"]).startswith("TAMIL NADU") ):
        return 'outsider-L'
    if ( str(x["LOCAL_PATIENT_IDENTIFIER"]).startswith("R_") ) and (  str(x['statename']).startswith("KA") or str(x["statename"]).startswith("TAMIL NADU")    ):
        return "outside-R"
    if  str(x["LOCAL_PATIENT_IDENTIFIER"]).startswith("L_")  and   str(x['statename']).startswith("KE")   :
        return "inside-L"
    if  str(x["LOCAL_PATIENT_IDENTIFIER"]).startswith("LR_")  and   str(x['statename']).startswith("KE")   :
        return "inside-R"
```
